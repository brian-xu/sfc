```python
import numpy as np
import matplotlib.pyplot as plt
from sfc import gen_curve
```

```python
plt.rcParams["figure.figsize"] = [10, 7]
plt.rcParams["figure.autolayout"] = True
```

```python
width = 40
height = 40
curve = gen_curve(width, height, "s_0")
```

```python
x = [0 for _ in range(width*height)]
y = [0 for _ in range(width*height)]
for i in range(height):
    for j in range(width):
        x[curve[i][j]] = j
        y[curve[i][j]] = i
```

```python
plt.plot(x, y)
plt.show()
```
